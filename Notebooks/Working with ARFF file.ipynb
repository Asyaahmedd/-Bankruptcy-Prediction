```python
# Import necessary libraries
from scipy.io import arff
import pandas as pd
```

```python
# Load ARFF file using the 'arff' library
data, meta = arff.loadarff('E:/Asya/polish+companies+bankruptcy+data/4year.arff')

# Convert the loaded data to a Pandas DataFrame
df = pd.DataFrame(data)

# Display the number of rows in the DataFrame
print(len(df))

# Display the first few rows of the DataFrame
df.head()
```

```python
# Rename the 'class' column to 'bankrupt'
df.rename(columns={'class': 'bankrupt'}, inplace=True)
# Convert values in the 'bankrupt' column from b'0' to False and b'1' to True
df['bankrupt'] = df['bankrupt'].replace({b'0': False, b'1': True})

# Display the count of unique values in the 'bankrupt' column (after conversion)
print(df["bankrupt"].value_counts())

```

```python
# Save the DataFrame to a CSV file
df.to_csv('bankruptcy_data.csv', index=False)

# Print a message to indicate that the CSV file has been saved
print("CSV file saved successfully.")
```
